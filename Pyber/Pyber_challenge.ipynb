```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
combined_data = pd.merge(city_data,ride_data,on="city")
# Display the data table for preview
combined_data
```

```python
#Calculated necessary values
grouped_data = combined_data.groupby(['city'])

driver_count = grouped_data.driver_count.mean()
average_fare = grouped_data.fare.mean()
total_rides = grouped_data.city.count()
city_type = grouped_data.type.unique()

#Creating detailed dataframe
dataDF = pd.DataFrame({"average_fare":average_fare,
                       "rides_per_city":total_rides,
                       "drivers_per_city":driver_count,
                        "city_type":city_type
                      })

#Remove brackets from city_type
dataDF['city_type'] = dataDF['city_type'].str.get(0)

#Check the dataframe
dataDF
```

## Bubble Plot of Ride Sharing Data

```python
#Color Dictionary for city_type
color_dict = {'Urban':'gold','Suburban':'lightskyblue','Rural':'lightcoral'}

#Plot the scatter with all parameters
plt.scatter(dataDF.rides_per_city,dataDF.average_fare,s=(dataDF.drivers_per_city)*11,alpha=.5,edgecolors='k',
           c=[color_dict[i] for i in dataDF.city_type ],label=['hello','world'])

#Label axis, title
plt.xlabel("Total Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title("Ride Sharing Data (2016)")

#Legend and extra note
plt.legend()
plt.text(42.5,42.5,'Note: Circle size correlates with drivers per city')
```

## Total Fares by City Type

```python
# New Df grouped by city_type
typeDF = combined_data.groupby(['type'])

# Build Pie Chart
avgfare_type = typeDF.fare.sum()
plt.pie(avgfare_type,labels=['Rural','Suburban','Urban'],colors=['lightcoral','lightskyblue','gold'],autopct="%1.1f%%")
# # Save Figure
```
